```python
x_train = x_train.drop('loan_id', axis=1)
x_test = x_test.drop('loan_id', axis=1)
# loan id is just an idenifier which has no meaning, if we
# it,the model may learn noise
# removing unnecessary columns from the model train and test
```

```python
#encoding the model x train and x test
# one hot encode  categorical  features

from sklearn.preprocessing import OneHotEncoder

categorical_cols = ["education","self_employed"]
ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_cat = ohe.fit_transform(x_train[categorical_cols])
x_test_cat = ohe.transform(x_test[categorical_cols])

# the drop is to remove the redundant columns to aviod dummy
# variable tarp
#sparse in default would store them as a numpy array,
# if false otll give the encoded data as a normal table of
#numbers instead compressing it
```

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = x_train.drop(columns=categorical_cols).columns
scaler = StandardScaler()
x_train_num = scaler.fit_transform(x_train[numeric_cols])
x_test_num = scaler.transform(x_test[numeric_cols])
```

```python
from sklearn.linear_model import LogisticRegression

x_train_cat_df = pd.DataFrame(x_train_cat, columns=ohe.get_feature_names_out(categorical_cols), index=x_train.index)
x_test_cat_df = pd.DataFrame(x_test_cat, columns=ohe.get_feature_names_out(categorical_cols), index=x_test.index)


x_test_num_df = pd.DataFrame(
    x_test_num,
    columns=numeric_cols,
    index=x_test.index
)

x_train_final = pd.concat([x_train_num_df, x_train_cat_df], axis=1)
x_test_final = pd.concat([x_test_num_df, x_test_cat_df], axis=1)

model = LogisticRegression(max_iter=100)
model.fit(x_train_final ,y_train)
```

```python
# evaluate accuracy
from sklearn.metrics import accuracy_score

y_pred=model.predict(x_test_final)
accuracy_score(y_test,y_pred)
```
